# Seaborn Visualization Library

Seaborn is a pyhton data visualization library based on matplotlib.It provides a high-level interface for drawing attractive 
and informative statisitical graphics. 

SEaborn helps you explore and understand the data.Its plotting fuctions opreate 
on dataframes and array containing whole datasets and internaly perform the necessary mapping and statisital aggregation to
produce informative plots.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#github-----> dataset1
dsiris=pd.read_csv('iris.csv')
print(dsiris)
print(dsiris.head())
sns.set(style="whitegrid")
ax=sns.stripplot(x='class',y='sepal length',data=dsiris)
plt.title('Graph')
plt.show()
```

```python
ax=sns.swarmplot(x='class',y='sepal length',data=dsiris)
plt.title('Graph')
plt.show()
```
